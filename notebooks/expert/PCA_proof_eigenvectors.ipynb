# Математическое обоснование
---

Рассмотрим данные в виде матрицы $X\in\mathbb{R}^{N\times d}$, центрированные по строкам (каждый объект имеет нулевое среднее по признакам). \
Обозначим строки $X$ как $x_i^\top\in\mathbb{R}^d$, $i=1,\dots,N$. Ковариационная матрица
$$
\Sigma = \frac1N \sum_{i=1}^N x_i\,x_i^\top
$$
симметрична и положительно полуопределена, поэтому все её собственные значения вещественны и неотрицательны.

---

### 1. Одномерная главная компонента

Мы хотим найти направление (единичный вектор) $w\in\mathbb{R}^d$, $\|w\|=1$, такое что проекция данных $z_i = w^\top x_i$ имеет максимальную дисперсию:
$$
\max_{\|w\|=1}\mathrm{Var}(z)
=
\max_{\|w\|=1}\frac1N\sum_{i=1}^N (w^\top x_i)^2
=
\max_{\|w\|=1}w^\top\!\Bigl(\tfrac1N\sum_{i=1}^N x_i x_i^\top\Bigr)w
=
\max_{\|w\|=1}w^\top\Sigma\,w.
$$

#### 1.1. Вариационная формулировка и метод Лагранжа

Задача
$$
\max_{w\in\mathbb{R}^d}w^\top\Sigma\,w
\quad\text{s.t.}\quad w^\top w = 1
$$
решается через лангранжиан
$$
\mathcal{L}(w,\lambda)
=
w^\top \Sigma\,w - \lambda\,(w^\top w - 1).
$$
Условие стационарности по $w$:
$$
\nabla_w \mathcal{L}
=
2\,\Sigma\,w - 2\,\lambda\,w = 0
\;\;\Longrightarrow\;\;
\Sigma\,w = \lambda\,w.
$$
Это уравнение собственных векторов. Значение $w^\top\Sigma\,w$ на решении равно
$$
w^\top\Sigma\,w = w^\top(\lambda w) = \lambda\,\|w\|^2 = \lambda.
$$
Чтобы максимизировать, выбираем наибольшее собственное значение $\lambda_1$ и соответствующий ему собственный вектор $v_1$.

> **Замечание о характере экстремума.**
> Для симметричных $\Sigma$ функция $f(w)=w^\top\Sigma\,w$ ограничена сверху и снизу на единичной сфере, и собственный вектор, соответствующий $\lambda_1$, даёт глобальный максимум.

---

### 2. Многомерная главная компонента ($k>1$)

Ищем $k$ ортонормированных направлений $w_1,\dots,w_k\in\mathbb R^d$, $w_i^\top w_j=\delta_{ij}$, которые вместе максимизируют сумму дисперсий проекций:
$$
\max_{\substack{W=[w_1,\dots,w_k]\\W^\top W=I_k}}
\sum_{j=1}^k \mathrm{Var}(w_j^\top x)
=
\max_{W^\top W=I_k}
\sum_{j=1}^k w_j^\top \Sigma\,w_j
=
\max_{W^\top W=I_k}\mathrm{tr}(W^\top\Sigma\,W).
$$

#### 2.1. Теорема Куранта–Фишера (вариационный принцип)

Для симметричной $\Sigma$ с упорядоченными собственными значениями
$$
\lambda_1 \ge \lambda_2 \ge \dots \ge \lambda_d \ge 0
$$
и соответствующими ортонормированными собственными векторами $v_1,\dots,v_d$ справедливо
$$
\max_{W^\top W=I_k} \mathrm{tr}(W^\top \Sigma\,W)
= \sum_{j=1}^k \lambda_j,
$$
и оптимальный $W$ задаётся первыми $k$ собственными векторами:
$$
W^* = [\,v_1,\;v_2,\;\dots,\;v_k\,].
$$

##### Идея доказательства
1. Любая $k$-мерная ортонормированная система столбцов $w_1,\dots,w_k$ представима через базис из $v_i$.
2. Тогда
   $$
   \mathrm{tr}(W^\top\Sigma\,W)
   = \sum_{j=1}^k w_j^\top\Sigma\,w_j
   = \sum_{j=1}^k \sum_{i=1}^d \lambda_i\,(v_i^\top w_j)^2.
   $$
3. Используя ортонормированность и неотрицательность $\lambda_i$, показывается, что максимальная сумма достигается при $(v_i^\top w_j)^2 = 1$ для $i=j\le k$ и нулях в остальных позициях.

---

### 3. Итог

- **Одномерно**: оптимальный $w$ — собственный вектор $\Sigma$ с наибольшим $\lambda$.
- **Многомерно**: оптимальное подпространство размером $k$ порождается первыми $k$ собственными векторами $\Sigma$.

Таким образом, PCA эквивалентна разложению $\Sigma = V\Lambda V^\top$ и отбору $k$ главных компонент $V_k = [v_1,\dots,v_k]$.
